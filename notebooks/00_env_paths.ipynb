```python
# =========================
# 00_env_paths.ipynb
# Fundação: Drive + paths + versões + seed
# =========================

# 1) Drive
from google.colab import drive
drive.mount('/content/drive')

# 2) Paths (ajuste a pasta do seu projeto)
from pathlib import Path

PROJECT_ROOT = Path("/content/drive/MyDrive/Mestrado")  # ajuste se necessário

RAW_PATH       = PROJECT_ROOT / "02-datasets" / "01-raw"
PROCESSED_PATH = PROJECT_ROOT / "02-datasets" / "02-processed"
FEATURES_PATH  = PROJECT_ROOT / "02-datasets" / "03-features"
MODELS_PATH    = PROJECT_ROOT / "03-models"
REPORTS_PATH   = PROJECT_ROOT / "04-reports"

for p in [RAW_PATH, PROCESSED_PATH, FEATURES_PATH, MODELS_PATH, REPORTS_PATH]:
    p.mkdir(parents=True, exist_ok=True)

print("PROJECT_ROOT:", PROJECT_ROOT)
print("RAW_PATH:", RAW_PATH)
print("PROCESSED_PATH:", PROCESSED_PATH)
print("FEATURES_PATH:", FEATURES_PATH)

# 3) Versões / dependências mínimas (não instala; apenas verifica)
import sys, platform
import pandas as pd
import numpy as np

try:
    import pyarrow as pa
    import pyarrow.parquet as pq
    ARROW_OK = True
except Exception as e:
    ARROW_OK = False
    print("pyarrow não disponível:", e)

print("\nAmbiente")
print("Python:", sys.version)
print("Platform:", platform.platform())
print("pandas:", pd.__version__)
print("numpy:", np.__version__)
print("pyarrow:", pa.__version__ if ARROW_OK else "N/A")

# 4) Seed + utilitários
import random, os
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)

def log(msg: str):
    print(f"[LOG] {msg}")

log("Ambiente inicializado.")
```
